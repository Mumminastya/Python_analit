```python
#Условие 1: Задача 1
import pandas as pd
import dateutil
import numpy as np

df = pd.read_csv('data/kc_final.csv')
df
```

```python
#1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
##Если реновации дома не было, то в новом признаке поставьте 0
def delta_renovated_function(row):
    # code goes here
    if (row['yr_renovated'] == 0):
        return 0
    else:
        return row['yr_renovated'] - row['yr_built']

df['delta_renovated'] = df.apply(delta_renovated_function, axis=1)
df.head()
```

```python
##1.3 Создайте признаки года продажи, месяца продажи
def date_function(row):
    return dateutil.parser.parse(row['date']).strftime("%Y")

def month_function(row):
    return dateutil.parser.parse(row['date']).strftime("%m")

df['sale_year'] = df.apply(date_function, axis=1)
df['sale_month'] = df.apply(month_function, axis=1)
df.head()
```

```python
#1.4 Удалите признаки date, zipcode, lat, long
df_s = df.drop(columns=['date', 'zipcode', 'lat', 'long'])
df_s.head()
```

```python
#2.1 Присоедините к таблице clients данные по домам через метод join
df.set_index('id').join(clients.set_index('house_id'), how='inner')
```

```python
#2.2 Присоедините к таблице clients данные по домам через метод merge
#Это нужно, чтобы понимать, какие дома покупались клиентами
#house_id - это индексы датафрейма с домами
pd.merge(df, clients, left_on='id', right_on='house_id', how='inner')
```

```python
#Условие 3: 3 задача Составьте несколько сводных таблиц
##3.1 Найдите среднюю стоимость домов в зависимости от количества спален Отсортируйте от меньшей стоимости к большей
df_pivot=pd.pivot_table(df,
               index=["bathrooms"],
              values=["price"])
df_pivot.reindex(df_pivot['price'].sort_values(ascending=True).index)
```

```python
##3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома
pd.pivot_table(df,
               index=["condition"],
              values=["price"],
            aggfunc=[np.min, np.max, np.mean],)
```

```python
##3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида
pd.pivot_table(df,
               index=["waterfront","view"],
              values=["id"],
            aggfunc=[len])
```

```python
##3.4 Каких домов в зависимости от этажности и количества спален больше?
pd.pivot_table(df,
               index=["bathrooms","floors"],
              values=["id"],
            aggfunc=[len])
```

```python
##3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома
pd.pivot_table(df,
               index=["condition","grade"],
              values=["price"],
             aggfunc={'price': np.median})
```
