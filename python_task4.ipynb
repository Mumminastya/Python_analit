```python
##1.2. Изучите стоимости недвижимости
df = pd.read_csv('data/kc-house-data.csv')
sns.kdeplot(data=df, y='price', fill=True)
```

```python
##1.4. Изучите распределение года постройки
sns.kdeplot(data=df['yr_built'])
```

```python
#Условие 2: 2 задача
##2.1. Изучите распределение домов от наличия вида на набережную Постройте график Сделайте выводы
df_waterfront = df.groupby(['waterfront']).size()
print(df_waterfront)
df_waterfront.plot(kind='bar')

##большая часть домов имеет вид на набережную
```

```python
#Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
#Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.
sns.relplot(data=df, x="yr_renovated", y="yr_built", hue="price")
#дата ремонта не оказывает существенного влияния на стоимость, важнее год постройки
```

```python
sns.lineplot(data=df, x="floors", y="price")
#наиболее дороги дома 2.5 и 3.5 этажа
```

```python
sns.histplot(data=df, x="bathrooms", hue="price")
#наиболее дороги дома 2.5 и 3.5 этажа
```

```python
df_rooms = pd.pivot_table(data=df,index='bedrooms',columns='bathrooms',values='price')
sns.heatmap(df_rooms,annot=True)
#наиболее дорогие квартиры с 6-7 спальнями и комнатами
```

```python
sns.jointplot(data=df, x="sqft_living", y="sqft_lot", hue="price")
##наиболее дороги квартиры с максимальной жилой плоащадью
```
